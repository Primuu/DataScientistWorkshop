```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
dataset = pd.read_csv('D:/Projekty/DataScientistWorkshop/lab_08/files/VLagun_Phys_Years.csv')
```

```python
dataset.head()
```

```python
strong = dataset[dataset.Years == 'Year_strong_wind']
```

```python
weak = dataset[dataset.Years == 'Year_weak_wind']
```

```python
plt.title('Strong vs Weak wind')
plt.xlabel('SS')
plt.ylabel('Depth')
plt.scatter(strong.SS, strong.Depth, color='red', label='Strong', alpha=0.3)
plt.scatter(weak.SS, weak.Depth, color='lime', label='Weak', alpha=0.3)
plt.legend()
plt.show()
```

```python
dataset.Years = [1 if i == 'Year_strong_wind' else 0 for i in dataset.Years]
```

```python
x = dataset.drop(['Years'], axis=1)
y = dataset.Years.values
```

```python
# Normalization
x = (x - np.min(x)) / (np.max(x) - np.min(x))
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
print('Naive Bayes score: ', nb.score(X_test, y_test))
```
