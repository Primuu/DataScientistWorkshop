```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
dataset = pd.read_csv('D:/Projekty/DataScientistWorkshop/lab_08/files/VLagun_Lab_Basins.csv')
```

```python
dataset.head()
```

```python
east = dataset[dataset.Basin == 'East']
```

```python
west = dataset[dataset.Basin == 'West']
```

```python
plt.title('East vs West')
plt.xlabel('Windspeedinsitu')
plt.ylabel('DOC')
plt.scatter(east.Windspeedinsitu, east.DOC, color='red', label='East', alpha=0.3)
plt.scatter(west.Windspeedinsitu, west.DOC, color='lime', label='West', alpha=0.3)
plt.legend()
plt.show()
```

```python
dataset.Basin = [1 if i == 'West' else 0 for i in dataset.Basin]
```

```python
x = dataset.drop(['Basin'], axis=1)
y = dataset.Basin.values
```

```python
# Normalization
x = (x - np.min(x)) / (np.max(x) - np.min(x))
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
print('Naive Bayes score: ', nb.score(X_test, y_test))
```
